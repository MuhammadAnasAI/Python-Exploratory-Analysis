***Feature Engineering***

```python
df['rating'].unique()
```

```python
bins = [1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
labels = ['Very Low', 'Low', 'Bad', 'Average', 'Good', 'Best', 'Batter']
df['Rating_Category'] = pd.cut(df['rating'], bins=bins, labels=labels)
```

```python
df.head(7)
```

```python
df['Rating_Category'].value_counts()
```

```python
df.groupby('campus_name')['Rating_Category'].value_counts().sort_values(ascending=False)
```

**Plotting Techniques**

```python
plt.figure(figsize=(16, 6))
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(x='Rating_Category', data=df)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.scatterplot(x='Rating_Category', y='students', data=df)
plt.show()
```
